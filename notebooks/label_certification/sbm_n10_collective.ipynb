```python
delta_l = [0, 0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.5, 1]
delta = delta_l*8
```

```python
df_header = ["model", "delta", "certified_acc_mean", "certified_acc_std", "certified_ratio_mean", "certified_ratio_std"]
df_s = df_sample[["model", "certified_acc_mean", "certified_acc_std", "certified_ratio_mean", "certified_ratio_std"]]
df_s = df_s.assign(delta=delta)
df_s.to_csv(sample_file, index=False)
```

```python
df_c = df_coll[["model", "certified_acc_mean", "certified_acc_std", "certified_ratio_mean", "certified_ratio_std"]]
df_c = df_c.assign(delta=delta)
df_c.to_csv(coll_file, index=False)
```

```python
sample_models = ["SGC","APPNP","GCN Skip-PC","MLP"]
coll_models = ["SGC_coll","APPNP_coll","GCN Skip-PC_coll","MLP_coll"]
sample_ratio_mean = []
coll_ratio_mean = []
sample_ratio_std = []
coll_ratio_std = []
for (sample_model,coll_model) in zip(sample_models,coll_models):
    df_s = df_sample.loc[df_sample['model']==sample_model]
    df_c = df_coll.loc[df_coll['model']==coll_model]
    sample_ratio_mean.append(df_s['certified_ratio_mean'].values.tolist())
    sample_ratio_std.append(df_s['certified_ratio_std'].values.tolist())
    coll_ratio_mean.append(df_c['certified_ratio_mean'].values.tolist())
    coll_ratio_std.append(df_c['certified_ratio_std'].values.tolist())
```
