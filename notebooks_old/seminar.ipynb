```python
from typing import List
```

```python
class Student:

    def __init__(self, identifier, r1, r2, r3):
        self.id = identifier
        self.is_chair = False
        self.is_vis = False
        self.is_oral = False
        self.r1 = r1
        self.r2 = r2
        self.r3 = r3
        self.chair = None
        self.vis = None
        self.oral = None

    def get_reviewers(self):
        return self.r1, self.r2, self.r3

    def set_reviewer_positions(self, chair, vis, oral) -> bool:
        if chair.is_chair or vis.is_vis or oral.is_oral:
            return False
        self.chair = chair
        self.chair.is_chair = True
        self.vis = vis
        self.vis.is_vis = True
        self.oral = oral
        self.oral.is_oral = True
        return True
```

```python
def print_solution(students):
    for student in students:
        chair = student.chair
        vis = student.vis
        oral = student.oral
        print(f"Student {student.id}: chair {chair.id}; vis {vis.id}; oral {oral.id}")
```

```python
students = []
students.append(Student(0, 8, 1, 7))
students.append(Student(1, 5, 0, 9))
students.append(Student(2, 4, 5, 6))
students.append(Student(3, 6, 2, 1))
students.append(Student(4, 9, 8, 0))
students.append(Student(5, 1, 3, 9))
students.append(Student(6, 0, 7, 10))
students.append(Student(7, 10, 6, 3))
students.append(Student(8, 7, 4, 2))
students.append(Student(9, 3, 2, 10))
students.append(Student(10, 5, 4, 8))
```

```python
def update_count(n_gen):
    n_gen = n_gen + 1
    if n_gen % 1_000_000 == 0:
        print(f"Generated {n_gen} solutions.")
    return n_gen

def explore_solution(depth, students: List[Student], n_gen, chair, vis, oral):
    if students[depth-1].set_reviewer_positions(
            students[chair], students[vis], students[oral]
        ):
        gen_solution(depth, students, n_gen)
        students[chair].is_chair = False
        students[vis].is_vis = False
        students[oral].is_oral = False
    else:
        n_gen = update_count(n_gen)


def gen_solution(depth: int, students: List[Student], n_gen=0):
    if depth == 11:
        print_solution(students)
        update_count(n_gen)
        return True
    student = students[depth]
    depth = depth + 1
    r1, r2, r3 = student.get_reviewers()
    explore_solution(depth, students, n_gen, r1, r2, r3)
    explore_solution(depth, students, n_gen, r1, r3, r2)
    explore_solution(depth, students, n_gen, r2, r1, r3)
    explore_solution(depth, students, n_gen, r2, r3, r1)
    explore_solution(depth, students, n_gen, r3, r2, r1)
    explore_solution(depth, students, n_gen, r3, r1, r2)
    return False
```
