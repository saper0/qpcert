```python
a = torch.tensor([1, 2, 3], requires_grad=True, dtype=torch.float64)
a.requires_grad_(False)
b = torch.tensor([1, 1, 1], requires_grad=True, dtype=torch.float64)
c = a @ b
l = c + 1
l.backward()
print(b.grad)
b.grad = None
l = c * 2
l.backward()
print(b.grad)
```
